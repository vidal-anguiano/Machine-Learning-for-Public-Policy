```python
import sqlite3
from sodapy import Socrata
import pandas as pd
import config
```

```python
sqlite_file = 'db.db'
conn = sqlite3.connect(sqlite_file)
c = conn.cursor()
```

```python
client = Socrata("data.cityofchicago.org",'config.key')
```

```python
g = 'cdmx-wzbz'
v = '7nii-7srd'
l = 't28b-ys7j'
graffiti = client.get(g,limit=10000)
vacant = client.get(v,limit=10000)
lights = client.get(l,limit=10000)
```

```python
vacant_df = pd.DataFrame.from_records(vacant)
lights_df = pd.DataFrame.from_records(lights)
graffiti_df = pd.DataFrame.from_records(graffiti)
vacant_cols = set(vacant_df.columns)
lights_cols = set(lights_df.columns)
graffiti_cols = set(graffiti_df.columns)
```
